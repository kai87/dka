```python
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
def pure_token_incentive_function1(
    quality, critical_review_fee, constructive_review_fee
):
    return -0.5 * np.tanh(8 * (quality - 0.5))


def pure_token_incentive_function2(
    quality, critical_review_fee, constructive_review_fee
):
    return (
        0.5
        * (critical_review_fee - constructive_review_fee)
        / (critical_review_fee + constructive_review_fee)
    )


def pure_token_incentive_function(
    quality, critical_review_fee, constructive_review_fee
):
    return -0.5 * np.tanh(8 * (quality - 0.5)) + 0.5 * (
        critical_review_fee - constructive_review_fee
    ) / (critical_review_fee + constructive_review_fee)


critical_labels = range(1, 10)
constructive_labels = range(1, 10)
qualities = [0.1, 0.5, 0.9]

data = [
    [
        [
            pure_token_incentive_function(quality, critical_fee, constructive_fee)
            for critical_fee in critical_labels
        ]
        for constructive_fee in constructive_labels
    ]
    for quality in qualities
]

fig, axs = plt.subplots(ncols=3)
fig.set_size_inches(15, 4)
fig.suptitle("Testing the token incentive function", fontsize=16)
for i in range(3):
    axs[i].set_title(f"paper quality = {qualities[i]}")
    sns.heatmap(
        data[i],
        ax=axs[i],
        xticklabels=critical_labels,
        yticklabels=constructive_labels,
        vmin=-1,
        vmax=1,
    )
    axs[i].set_xlabel("critical review fee")
    axs[i].set_ylabel("constructive review fee")


data = [
    [
        [
            pure_token_incentive_function1(quality, critical_fee, constructive_fee)
            for critical_fee in critical_labels
        ]
        for constructive_fee in constructive_labels
    ]
    for quality in qualities
]

fig, axs = plt.subplots(ncols=3)
fig.set_size_inches(15, 4)
fig.suptitle("Testing -0.5 * np.tanh(8 * (quality - 0.5))", fontsize=16)

for i in range(3):
    axs[i].set_title(f"paper quality = {qualities[i]}")
    sns.heatmap(
        data[i],
        ax=axs[i],
        xticklabels=critical_labels,
        yticklabels=constructive_labels,
        vmin=-1,
        vmax=1,
    )
    axs[i].set_xlabel("critical review fee")
    axs[i].set_ylabel("constructive review fee")


data = [
    [
        [
            pure_token_incentive_function2(quality, critical_fee, constructive_fee)
            for critical_fee in critical_labels
        ]
        for constructive_fee in constructive_labels
    ]
    for quality in qualities
]

fig, axs = plt.subplots(ncols=3)
fig.set_size_inches(15, 4)
fig.suptitle(
    "Testing 0.5 * (critical_review_fee - constructive_review_fee) / (critical_review_fee + constructive_review_fee)",
    fontsize=16,
)

for i in range(3):
    axs[i].set_title(f"paper quality = {qualities[i]}")
    sns.heatmap(
        data[i],
        ax=axs[i],
        xticklabels=critical_labels,
        yticklabels=constructive_labels,
        vmin=-1,
        vmax=1,
    )
    axs[i].set_xlabel("critical review fee")
    axs[i].set_ylabel("constructive review fee")
```
