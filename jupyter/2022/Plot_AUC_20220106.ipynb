```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Acceptance rate 20%
dfs_20 = [pd.read_csv('./AUC_results/simulation_result_alpha_045.csv'), pd.read_csv('./AUC_results/simulation_result_beta_045.csv')]

columns_20 = [
    "2014-0.2",
    "2015-0.2",
    "2016-0.2",
    "2017-0.2",
    "2018-0.2"
]

labels_20 = [
    "KDD 2014, 20% Acceptance Rate",
    "KDD 2015, 20% Acceptance Rate",
    "KDD 2016, 20% Acceptance Rate",
    "KDD 2017, 20% Acceptance Rate",
    "KDD 2018, 20% Acceptance Rate",
]

# Acceptance rate 40%
dfs_40 = [pd.read_csv('./AUC_results/simulation_result_alpha_036.csv'), pd.read_csv('./AUC_results/simulation_result_beta_036.csv')]

columns_40 = [
    "2014-0.4",
    "2015-0.4",
    "2016-0.4",
    "2017-0.4",
    "2018-0.4"
]

labels_40 = [
    "KDD 2014, 40% Acceptance Rate",
    "KDD 2015, 40% Acceptance Rate",
    "KDD 2016, 40% Acceptance Rate",
    "KDD 2017, 40% Acceptance Rate",
    "KDD 2018, 40% Acceptance Rate",
]
```

```python
xs_20 = []
ys_20 = []
hues_20 = []
plot_dfs_20 = []
    
for i in range(2):
    xs = []
    ys = []
    hues = []

    for index in range(5):
        if i == 0:
            xs.extend(list(dfs_20[i]['a mean']))
        if i == 1:
            xs.extend(list(dfs_20[i]['b mean']))
            
        ys.extend(list(dfs_20[i][columns_20[index]]))
        hues.extend([labels_20[index]] * len(list(dfs_20[i][columns_20[index]])))    
    
    plot_df = pd.DataFrame(
    data = {
        'Alpha/Beta': xs,
        'AUC': ys,
        'hue': hues,
    }
    )
    
    xs_20.append(xs)
    ys_20.append(ys)
    hues_20.append(hues)
    plot_dfs_20.append(plot_df)

# this is the gap between error bars
gap = 0
    
plot_dfs_20[0]["Alpha/Beta"] = plot_dfs_20[0].apply(lambda x:  x['Alpha/Beta']+gap*(int(x['hue'][4:8])-2016), axis = 1)
plot_dfs_20[1]["Alpha/Beta"] = plot_dfs_20[1].apply(lambda x:  x['Alpha/Beta']+gap*(int(x['hue'][4:8])-2016), axis = 1)

fig, axes = plt.subplots(1, 2, figsize=(18, 5))




sns.lineplot(ax = axes[0], data = plot_dfs_20[0], x = "Alpha/Beta", y = "AUC", hue = "hue", err_style = "bars", palette=['red', 'darkorange', 'grey', 'royalblue', 'purple'], marker="o", markersize=7, ci="sd", err_kws={"capsize":3})
sns.lineplot(ax = axes[1], data = plot_dfs_20[1], x = "Alpha/Beta", y = "AUC", hue = "hue", err_style = "bars", palette=['red', 'darkorange', 'grey', 'royalblue', 'purple'], marker="o", markersize=7, ci="sd", err_kws={"capsize":3})

axes[0].set_xlabel(r"Token Attractiveness Score ($\alpha$)", fontsize=12)
axes[1].set_xlabel(r"Reputation Attractiveness Score ($\beta$)", fontsize=12)
axes[0].legend(loc = "lower left")
axes[1].legend(loc = "lower right")

for ax in axes:
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles=handles[1:], labels=labels[1:])
    ax.set_ylim([0.5, 1])
    ax.set_ylabel(r"Area Under The Curve (AUC)", fontsize = 12)
    ax.grid(color = "lightgrey", linestyle = "--", linewidth = 1)
    ax.tick_params(axis = 'x', labelsize = 12)
    ax.tick_params(axis = 'y', labelsize = 12)
```
