```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
kdd_papers = pd.read_csv(r"results/kdd_papers.csv", index_col=0)
print(set(kdd_papers.Year))
kdd_papers = kdd_papers[kdd_papers.Year <= 2015]
kdd_papers.CitationCount.hist(bins=100, density=True)
plt.title("Citation density for KDD 2010-2015")
plt.xlabel("Citation count")
plt.ylabel("Density")
plt.savefig("citation_count.png")
```
